```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Figura cubo
# Definir valores
Xmin, Xmax = -2, 2
Ymin, Ymax = -2, 2
Zmin, Zmax = -2, 2

# Definir valores animaciones
frames = 1000
t_pause = 0.001
t_pauseInicio = 0.1

# Definir angulo
alpha = 90
beta = 90
theta = 90

# Vertices
a = [0, 0, 0]
b = [1, 0, 0]
c = [1, 1, 0]
d = [0, 1, 0]
e = [0, 0, 1]
f = [1, 0, 1]
g = [1, 1, 1]
h = [0, 1, 1]

# Matriz de puntos
x=[a[0], b[0], c[0], d[0], a[0], e[0], f[0], g[0], h[0], e[0], h[0], d[0], h[0], d[0], h[0], g[0], c[0], b[0], f[0]];
y=[a[1], b[1], c[1], d[1], a[1], e[1], f[1], g[1], h[1], e[1], h[1], d[1], h[1], d[1], h[1], g[1], c[1], b[1], f[1]];
z=[a[2], b[2], c[2], d[2], a[2], e[2], f[2], g[2], h[2], e[2], h[2], d[2], h[2], d[2], h[2], g[2], c[2], b[2], f[2]];

# Crear la figura y los ejes 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Dibujar las líneas del cubo
ax.plot(x, y, z, color='b')

# Etiquetas de los ejes
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

# Mostrar la gráfica
plt.show()
```
